```python
import pandas as pd
r_cols = ['user_id','movie_id','rating']
ratings = pd.read_csv('u.data', sep='\t', names = r_cols, usecols=range(3), encoding='ISO-8859-1')

movie_cols =['movie_id','title']
movies = pd.read_csv('u.item', sep='|', names = movie_cols, usecols=range(2), encoding='ISO-8859-1')
ratings = pd.merge(movies, ratings)
ratings.head(15)
```

```python
userRatings = ratings.pivot_table(index=['user_id'], columns=['title'], values=['rating'])
userRatings.head()
```

```python
#correlation between every column pair found in matrix
corrMatrix = userRatings.corr()
corrMatrix.head()
```

```python
#inorder to avoid spurious results, we want to consider corr scores for movies backed up by atleast a 100 users
corrmatrix = userRatings.corr(method='pearson', min_periods=100)
corrmatrix.head()
```

```python
myRate = userRatings.loc[0].dropna()
myRate.head()
```

```python
# we will be recommending to this user lets say its me , I am going to go through all the movies I rated one time and build up a list of recommendations based on movies similar to the ones I rated.
similarCandidates = pd.Series(dtype='float64')
for i in range(0, len(myRate.index)):
    print('finding similar candidates from {}'. format(myRate.index[i]))
    sims = corrMatrix[myRate.index[i]].dropna()
    sims = sims.map(lambda x:x*myRate[i]) # scaling the similarity by how well I liked the movie
    similarCandidates = similarCandidates.append(sims)
similarCandidates.sort_values(ascending=False)
similarCandidates.head(10)
```
